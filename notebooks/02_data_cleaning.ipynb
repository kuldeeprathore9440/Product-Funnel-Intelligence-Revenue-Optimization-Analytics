```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
sns.set_style("whitegrid")
```

```python
users = pd.read_csv(r"C:\Users\kuldiep\Downloads\Data\users.csv")
events = pd.read_csv(r"C:\Users\kuldiep\Downloads\Data\events.csv")
orders = pd.read_csv(r"C:\Users\kuldiep\Downloads\Data\orders.csv")
```

```python
users["signup_date"] = pd.to_datetime(users["signup_date"], errors = "coerce")
events["timestamp"] = pd.to_datetime(events["timestamp"], errors = "coerce")
orders["order_date"] = pd.to_datetime(orders["order_date"], errors = "coerce")
orders["order_value"] = pd.to_numeric(orders["order_value"], errors = "coerce")
```

```python
users["country"] = users["country"].astype(str).str.strip().str.title()
users["device"] = users["device"].astype(str).str.strip().str.title()
users["channel"] = users["channel"].astype(str).str.strip().str.title()

events["event"] = events["event"].astype(str).str.strip().str.lower()
```

```python
users = users.drop_duplicates()
events = events.drop_duplicates()
orders = orders.drop_duplicates()
```

```python
users = users[users["user_id"].notna()]
events = events[events["user_id"].notna()]
orders = orders[orders["user_id"].notna()]
```

```python
valid_events = ["visit", "signup", "activate", "add_to_cart", "purchase"]
events = events[events["event"].isin(valid_events)]
```

```python
orders = orders[orders["order_id"].notna()]
orders = orders[orders["order_id"] > 0]
```

```python
users = users.drop_duplicates(subset = ["user_id"])
```

```python
valid_user_ids = set(users["user_id"])

events = events[events["user_id"].isin(valid_user_ids)]
orders = orders[orders["user_id"].isin(valid_user_ids)]
```

```python
users.to_csv("users_clean.csv", index = False)
events.to_csv("events_clean.csv", index = False)
orders.to_csv("orders_clean.csv", index = False)
```
